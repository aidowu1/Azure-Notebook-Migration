### An exploration of Logistic Regression activation fuctions

```python
import numpy as np
import matplotlib.pylab as pl

%matplotlib inline

# Define the softmax function
def softmax(z):
    return np.exp(z) / np.sum(np.exp(z))

def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

### Logistic Regression Profile

```python
def plotActivationFuncs(x, y, labels):
    pl.plot(x, y)
    pl.xlabel(labels[0])
    pl.ylabel(labels[1])
    pl.show()

x = np.arange(-2, 2)
#y = softmax(x)
y = sigmoid(x)
plotActivationFuncs(x, y, labels = ("x", "Sigmoid Scores"))
```

### Softmax Profile

```python
y_softmax = softmax(x)
plotActivationFuncs(x, y_softmax, labels = ("x", "Softmax Scores"))
```

```python
z_D = np.exp(-9)
z_E = np.exp(1)
z_sum = z_D + z_E
z_softmax = [z_D/z_sum, z_E/z_sum]
print("z_D = {}".format(z_D))
print("z_E = {}".format(z_E))
print("z_softmax = {}".format(z_softmax))
```
