### Install Quandl, we will use it to get stock data from Google Finance

```python
#! pip install quandl 
```

```python
import quandl
import os
import pandas as pd
stock_data_cache = 'aapl.pl'
def getStockData():
    if not os.path.exists(stock_data_store):
        aapl = quandl.get("WIKI/AAPL", start_date="2006-10-01", end_date="2012-01-01")
        aapl.to_pickle(stock_data_store)
    else:
        aapl = pd.read_pickle(stock_data_store)
    return aapl

aapl = getStockData()
```

### We can save the data to csv

```python
stock_data_csv_path = 'aapl.csv'
aapl.to_csv(stock_data_csv_path)
```

### Let us plot the 'Close' price of the time-series data

```python
# Import Matplotlib's `pyplot` module as `plt`
import matplotlib.pyplot as plt

%matplotlib inline

def plotStockData(column = 'Close'):
   # Plot the closing prices for `aapl`
    aapl[column].plot(grid=True)

    # Show the plot
    plt.show()

plotStockData()
```

### Compute the daily percentage change of the 'Close' price

```python
def computePctChange(column = 'Close'):
    aapl['pct_change_{0}'.format(column)] = aapl[column].pct_change()   

computePctChange()
plotStockData('pct_change_Close')
```
